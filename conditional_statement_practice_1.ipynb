```python
### weight:40
# use coditions:
# weight<35-"underweight"
# weight<45-"normalweight"
# else:-"overweight"

# take the input from input
weight=int(input("enter the weight:"))
if weight<=35:
    print("under weight")
elif weight<=45:
    print("normal weight")
else:
    print("over weight")

```
