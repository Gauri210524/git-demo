Q1. Print numbers from 1 to 20 using while loop.

Approach:
- Initialize a variable
- Use while condition
- Increment inside loop

```python
#write answer here
a=1
while a<=20:
  print(a)
  a+=1
```

Q4. Find sum of numbers from 1 to n.

Approach:
- Take input n
- Add numbers in loop

```python
#write answer here
n=int(input("enter any number"))
count=1
sum=0
while count<=n:
  sum=sum+count
  count=count+1
  print("sum of numbers from 1 to ",n, "is",sum)
```

Q5. Reverse a number using while loop.

Approach:
- Extract last digit
- Build reversed number

```python
#write answer here

var= int(input())

new = var

res=0

while var>0:
  temp = var % 10
  res = res*10 + temp
  var = var // 10

print(res)


if new==res:
  print("Palindrome")
else:
  print("Not Palindrome")


```

Q6. Count digits in a number.

Approach:
- Divide number by 10 repeatedly
- Count iterations

```python
#write answer here
n=int(input("enter any number"))
count=0
while count<n:
  n=n//10
  count=count+1
print("number of digits",count)
```
