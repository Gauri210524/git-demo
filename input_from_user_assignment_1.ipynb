```python
# find the bill with gst and dicount of amount
# take the input from user
bill=int(input("Enter the bill amount:"))
gst_tax=.18
sgst=0.12
discount=0.20
final_bill=bill+bill*gst_tax+bill*sgst-bill*discount
print("The final bill amount is:",final_bill)
```
